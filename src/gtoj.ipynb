```python
"""
Creating same plots using giotto's mapper.
"""

# Data handling
import numpy as np
import pandas as pd

# Data vizualization
from gtda.plotting import plot_point_cloud

from gtda.mapper import (
    CubicalCover,
    make_mapper_pipeline,
    Projection,
    plot_static_mapper_graph,
    plot_interactive_mapper_graph,
    MapperInteractivePlotter
)

from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
from gtda.mapper.cluster import FirstSimpleGap


from main import GenerateVectors

from sklearn import datasets


DengueDataCSV = '../data/strict-data/2010DengueData.csv'
DengueDataCSV2 = '../data/created-data/north_increase/factor/north_factor=2.csv'

DengueDataFrame = pd.read_csv(DengueDataCSV2)

VectorList, LabelsList = GenerateVectors(DengueDataFrame)
NormalizedVectorList = normalize(VectorList, norm='l2', axis=0)
print(NormalizedVectorList)



# data, _ = datasets.make_circles(n_samples=5000, noise=0.05, factor=0.3, random_state=42)
# print(data)
# plot_point_cloud(np.array(VectorList))

filter_func = Projection(columns=[0,1,2,3])
cover = CubicalCover(n_intervals=10, overlap_frac=0.15)
clusterer = DBSCAN()

n_jobs = 1

pipe = make_mapper_pipeline(
    filter_func=filter_func,
    cover=cover,
    clusterer=clusterer,
    verbose=False,
    n_jobs=n_jobs
)


fig = plot_static_mapper_graph(pipe, NormalizedVectorList, layout_dim=3)
fig.show(config={'scollZoom': True})
```
